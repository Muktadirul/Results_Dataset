```python
import os
import pandas as pd
from glob import glob
```

```python
Excel_files_list = []
path_lists = ['resultAstral/','resultPhylonet/']
for path_list in path_lists:
    for i in os.listdir(path_list):
        for j in os.listdir(path_list+i+'/'):
            for k in os.listdir(path_list+i+'/'+j+'/'):
                a = glob(path_list+i+'/'+j+'/'+k+'/*.csv')
                if a:
                    Excel_files_list.append(a)
count = 0     
csv_files_list = []
for i in range(0,len(Excel_files_list)):
    for j in range(0,len(Excel_files_list[i])):
        path = Excel_files_list[i][j]
        replace_last = path.replace(path.split('/')[-1],'')
        fn_path = glob(replace_last.replace('Merged_data','FNRATE_*'))
        files = glob(fn_path[0]+'/*[0-9]')
        for index in range(0,len(files)):
            os.system('cat '+files[index]+' > '+files[index]+'.txt')
            file_reader = open(files[0],'r')
            data = file_reader.readlines()
            count +=1
            print('===============================================================================\n')
            print(files[index])
            print('===============================================================================\n')
            
            fnrate_array = []
            try:
                for fn_rate in data:
                    fnrate_array.append(fn_rate.split(',')[-2])

            except Exception as ex:
                type_name = files[index].split('/')[-1]
                Replace_pattern = files[index].split('/')[-2]+'/'+files[index].split('/')[-1]
                if 'exact' in type_name:
                    if not os.path.exists(files[index].replace(Replace_pattern,'00MAIN/')):
                        os.mkdir(files[index].replace(Replace_pattern,'00MAIN/'))
                    if 'FN' in type_name and '_exact' in type_name:
                        df = pd.DataFrame({'FNRATE_EXACT':fnrate_array})
                        df.to_csv(files[index].replace(Replace_pattern,'00MAIN/'+type_name+'.csv'))
                    else:
                        df = pd.DataFrame({'FNRATE':fnrate_array})
                        df.to_csv(files[index].replace(Replace_pattern,'00MAIN/'+type_name+'.csv'))
                else:
                    if not os.path.exists(files[index].replace(Replace_pattern,'00MAIN/')):
                        os.mkdir(files[index].replace(Replace_pattern,'00MAIN/'))
                        
                    if 'FN' in type_name and '_exact' in type_name:
                        df = pd.DataFrame({'FNRATE_EXACT':fnrate_array})
                        df.to_csv(files[index].replace(Replace_pattern,'00MAIN/'+type_name+'.csv'))
                    else:
                        df = pd.DataFrame({'FNRATE':fnrate_array})
                        df.to_csv(files[index].replace(Replace_pattern,'00MAIN/'+type_name+'.csv'))
                    
            print('=========================================='+str(count)+'=====================================\n')
```

```python
tr_cnt = 0
normal_cnt = 0
exact_cnt =0

exact_list = []
true_list = []
normal_list = []

for i in range(0,len(Excel_files_list)):
    for j in range(0,len(Excel_files_list[i])):
        path = Excel_files_list[i][j]
        type_name = path.split('/')[-1]
        Replace_pattern = path.split('/')[-2]+'/'+path.split('/')[-1]
        if 'exact' in path:
            exact_cnt+=1
            if not os.path.exists(path.replace(Replace_pattern,'00MAIN/')):
                os.mkdir(path.replace(Replace_pattern,'00MAIN/'))
            out_path = path.replace(Replace_pattern,'00MAIN/')+type_name
            os.system('cat '+path+' > '+out_path)
            exact_list.append(out_path)

        elif 'True' in path:
            tr_cnt+=1
            if not os.path.exists(path.replace(Replace_pattern,'03True/')):
                os.mkdir(path.replace(Replace_pattern,'03True/'))
            out_path = path.replace(Replace_pattern,'03True/')+type_name
            os.system('cat '+path+' > '+out_path)
            true_list.append(out_path)
        
        else:
            normal_cnt+=1
            if not os.path.exists(path.replace(Replace_pattern,'00MAIN/')):
                os.mkdir(path.replace(Replace_pattern,'00MAIN/'))
            out_path = path.replace(Replace_pattern,'00MAIN/')+type_name
            os.system('cat '+path+' > '+out_path)
            normal_list.append(out_path)
            
print(tr_cnt,normal_cnt,exact_cnt)
```

```python
SHEET_NAMES = []
Excel_files_list=[]
# path_lists = ['resultPhylonet/']
path_lists = ['resultAstral/']
for path_list in path_lists:
    for i in os.listdir(path_list):
        for j in os.listdir(path_list+i+'/'):
            for k in os.listdir(path_list+i+'/'+j+'/'):
                a = glob(path_list+i+'/'+j+'/'+k+'/*.csv')
                if a:
                    Excel_files_list.append(a)
                    
writer = pd.ExcelWriter('Allresults_Astral.xlsx', engine='xlsxwriter')
List_of_result_dataframe = {}
for number_s in range(0,len(Excel_files_list)):
    files_list1 = Excel_files_list[number_s]    
    files_list = []
    for f_list in files_list1:
        if '00MAIN' in f_list:
            files_list.append(f_list)
        
    column_names = []
    for files in files_list:
        name = (files.split('/')[-1]).replace('.csv','')
        
        if 'FN' in name and 'exact' in name:
            column_names.append('FNRATE_EXACT')
        elif 'FN' in name:
            column_names.append('FNRATE')
        else:
            column_names.append(name)

    merged_data = {}
    store_file_sheet_name =''
    data = pd.DataFrame(columns=column_names)
    
    for files in files_list:
        dt = pd.read_csv(files)
        dt_col = list(dt.columns)
        
        data['TAXON'] = files.split('/')[-4]
        data['MODEL_CONDITION'] = files.split('/')[-3]
        store_file_sheet_name = files.split('/')[-3]
        if str(dt_col[0]) in column_names:
            data[dt_col[0]] = dt[dt_col[0]]
        elif str(dt_col[1]) in column_names:
            data[dt_col[1]] = dt[dt_col[1]]
        data['GENE'] = data.index
    data = data.rename(columns={'FNRATE': 'FNRATE_ASTRAL','FNRATE_EXACT': 'FNRATE_EXACT_ASTRAL'})
    data.to_excel(writer,sheet_name=store_file_sheet_name)
    SHEET_NAMES.append(store_file_sheet_name)
writer.save()
```

```python
SHEET_NAMES = list(set(SHEET_NAMES))
del(SHEET_NAMES[0])
len(SHEET_NAMES)
```

```python

writer1 = pd.ExcelWriter('EL.xlsx', engine='xlsxwriter')
writer2 = pd.ExcelWriter('QS.xlsx', engine='xlsxwriter')

for i in SHEET_NAMES:
    df1 = pd.read_excel('Allresults_Phylonet.xlsx',sheet_name=i)
    df2 = pd.read_excel('Allresults_Astral.xlsx',sheet_name=i)
    df3 = df1.join(df2)
    columns_list = df3.columns
    EL_COLUMNS_LIST = []
    QS_COLUMNS_LIST = []
    for col in columns_list:
        if 'EL_' in col:
            EL_COLUMNS_LIST.append(col)
        elif 'QS_' in col:
            QS_COLUMNS_LIST.append(col)
        else:
            QS_COLUMNS_LIST.append(col)
            EL_COLUMNS_LIST.append(col)
            
    
    EL_DF = df3.loc[:,EL_COLUMNS_LIST] 
    QS_DF = df3.loc[:,QS_COLUMNS_LIST]
    
    EL_DF.to_excel(writer1,sheet_name=i)
    QS_DF.to_excel(writer2,sheet_name=i)
    
writer1.save()
writer2.save()
    
```

```python
# SPLITING EL WITH FN RATE
for i in SHEET_NAMES:
    EL_DF = pd.read_excel('EL.xlsx',sheet_name=i)
    columns_list = EL_DF.columns
    EL_COLUMNS_LIST = []
    FN_COLUMNS_LIST = []
    for col in columns_list:
        if 'EL_' in col:
            EL_COLUMNS_LIST.append(col)
        elif 'FNRATE' in col:
            FN_COLUMNS_LIST.append(col)
    for EL in EL_COLUMNS_LIST:
        for FN in FN_COLUMNS_LIST:
            LOCK_LIST = [EL,FN,'TAXON','MODEL_CONDITION','GENE']
            SPLIT_UP_EL = EL_DF.loc[:,LOCK_LIST]
            SPLIT_UP_EL = SPLIT_UP_EL[SPLIT_UP_EL.duplicated([EL], keep=False)]
            if not SPLIT_UP_EL.empty:
                SPLIT_WRITER = pd.ExcelWriter('EL/'+str(i+'_'+EL+'_'+FN)+'.xlsx', engine='xlsxwriter')
                SPLIT_UP_EL.to_excel(SPLIT_WRITER,sheet_name=i,index=False)
                SPLIT_WRITER.save()
```

```python
# SPLITING QS WITH FN RATE
for i in SHEET_NAMES:
    QS_DF = pd.read_excel('QS.xlsx',sheet_name=i)
    columns_list = QS_DF.columns
    QS_COLUMNS_LIST = []
    FN_COLUMNS_LIST = []
    for col in columns_list:
        if 'QS_' in col:
            QS_COLUMNS_LIST.append(col)
        elif 'FNRATE' in col:
            FN_COLUMNS_LIST.append(col)
    for QS in QS_COLUMNS_LIST:
        for FN in FN_COLUMNS_LIST:
            LOCK_LIST = [QS,FN,'TAXON','MODEL_CONDITION','GENE']
            SPLIT_UP_QS = QS_DF.loc[:,LOCK_LIST]
            SPLIT_UP_QS['MODEL_CONDITION'] = i
            SPLIT_UP_QS = SPLIT_UP_QS[SPLIT_UP_QS.duplicated([QS], keep=False)]
            if not SPLIT_UP_QS.empty:
                SPLIT_WRITER = pd.ExcelWriter('QS/'+str(i+'_'+QS+'_'+FN)+'.xlsx', engine='xlsxwriter')
                SPLIT_UP_QS.to_excel(SPLIT_WRITER,sheet_name=i,index=False)
                SPLIT_WRITER.save()
```

```python
QS_FNRATE_FILES = glob('QS/*.xlsx')
qs_cnt =0
file_writer = open('REPORT/QS_REPORT.txt','a+')

for i in QS_FNRATE_FILES:
    QS_FNRATE_DF_READ = pd.read_excel(i, ignore_index = True)
    QS_FNRATE_DF_columns =[]
    QS_FNRATE_DF_columns = QS_FNRATE_DF_READ.columns
    for col in QS_FNRATE_DF_columns:
        if 'QS_' in col:
            GRP_BY_NAME = col
        elif 'FNRATE' in col:
            FN_TYPE = col
    QS_FNRATE_DF_READ_GROUPED = QS_FNRATE_DF_READ.groupby(GRP_BY_NAME)
    QS_GRP_LIST = []
    for name,group in QS_FNRATE_DF_READ_GROUPED:
        QS_GRP_LIST.append(group)
        
    for j in range(0,len(QS_GRP_LIST)):
        FN_RATE_LIST = []
        QS_SCORE_LIST = []
        MODEL_LIST = []
        TAXON_LIST = []
        GENE_LIST = []
        if len(list(QS_GRP_LIST[j][FN_TYPE])) == len(set(list(QS_GRP_LIST[j][FN_TYPE]))):
            FN_RATE_LIST = QS_GRP_LIST[j][FN_TYPE]
            QS_SCORE_LIST = QS_GRP_LIST[j][GRP_BY_NAME]
            MODEL_LIST = QS_GRP_LIST[j]['MODEL_CONDITION']
            TAXON_LIST = QS_GRP_LIST[j]['TAXON']
            GENE_LIST = QS_GRP_LIST[j]['GENE']

            WRITE_DF = pd.DataFrame({
                'FN_RATE':FN_RATE_LIST,
                'QS_SCORE':QS_SCORE_LIST,
                'MODEL':MODEL_LIST,
                'Texon':TAXON_LIST,
                'GENE':GENE_LIST})
            WRITE_DF=WRITE_DF.reset_index(drop=True)
            file_writer.write('===============================================================================\n')
            str_dt1=str(FN_TYPE)+'\t QS SCORE\t MODEL_CONDITION \t\t\t Texon\t GENE No \n'
            file_writer.write(str_dt1)
            file_writer.write('===============================================================================\n')
            for k in range(0,len(WRITE_DF['FN_RATE'])):
                str_data = str(WRITE_DF['FN_RATE'][k])+'\t\t\t\t'+str(WRITE_DF['QS_SCORE'][k])+'\t\t'+str(WRITE_DF['MODEL'][k])+'\t\t'+str(WRITE_DF['Texon'][k])+'\t \t'+str(WRITE_DF['GENE'][k]+1)+'\n'
                file_writer.write(str_data)
            file_writer.write('===============================================================================\n\n')
            qs_cnt+=1
print(qs_cnt)
file_writer.close()
```
