```python
import pandas as pd
from glob import glob
```

```python
def data_merge(paths):
    data = []
    for path in paths:
        data.append(pd.read_csv(path,index_col=None, header=0))
    return data
```

```python
path = '/home/raian/Desktop/automate-score-el-qs-fn-std-stderr/Final_Result/11-texon/estimated_*genes_weakILS/'
Estmeted_EL = glob(path+'Result_of__Estimated__EL__*.csv')
Estmeted_FN = glob(path+'Result_of__Estimated__FN__*.csv')
Estmeted_QS = glob(path+'Result_of__Estimated__QS__*.csv')

EL_frame = pd.concat(data_merge(Estmeted_EL))
FN_frame = pd.concat(data_merge(Estmeted_FN))
QS_frame = pd.concat(data_merge(Estmeted_QS))

EL_frame['Score'] = EL_frame.apply(lambda row: str(row['Score']), axis=1)
EL_frame['stderr'] = EL_frame.apply(lambda row: str(row['stderr']), axis=1)

FN_frame['Score'] = FN_frame.apply(lambda row: str(row['Score']), axis=1)
FN_frame['stderr'] = FN_frame.apply(lambda row: str(row['stderr']), axis=1)

QS_frame['Score'] = QS_frame.apply(lambda row: str(row['Score']), axis=1)
QS_frame['stderr'] = QS_frame.apply(lambda row: str(row['stderr']), axis=1)

path1 = '/home/raian/Desktop/automate-score-el-qs-fn-std-stderr/Final_Result/11-texon/estimated_'
writer = pd.ExcelWriter(path1+'EL_QS_FN_weakILS.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
EL_frame.to_excel(writer, sheet_name='EL')
QS_frame.to_excel(writer, sheet_name='QS')
FN_frame.to_excel(writer, sheet_name='FN')

# # Close the Pandas Excel writer and output the Excel file.
writer.save()
```
