```python
import os
import pandas as pd
from glob import glob
```

```python
Excel_files_list = []
path_lists = ['resultAstral/','resultPhylonet/']
for path_list in path_lists:
    for i in os.listdir(path_list):
        for j in os.listdir(path_list+i+'/'):
            for k in os.listdir(path_list+i+'/'+j+'/'):
                a = glob(path_list+i+'/'+j+'/'+k+'/*.csv')
                if a:
                    Excel_files_list.append(a)
count = 0     
csv_files_list = []
for i in range(0,len(Excel_files_list)):
    for j in range(0,len(Excel_files_list[i])):
        path = Excel_files_list[i][j]
        replace_last = path.replace(path.split('/')[-1],'')
        fn_path = glob(replace_last.replace('Merged_data','FNRATE_*'))
        files = glob(fn_path[0]+'/*[0-9]')
        for index in range(0,len(files)):
            os.system('cat '+files[index]+' > '+files[index]+'.txt')
            file_reader = open(files[0],'r')
            data = file_reader.readlines()
            count +=1
            print('===============================================================================\n')
            print(files[index])
            print('===============================================================================\n')
            
            fnrate_array = []
            try:
                for fn_rate in data:
                    fnrate_array.append(fn_rate.split(',')[-2])

            except Exception as ex:
                type_name = files[index].split('/')[-1]
                Replace_pattern = files[index].split('/')[-2]+'/'+files[index].split('/')[-1]
                if 'exact' in type_name:
                    if not os.path.exists(files[index].replace(Replace_pattern,'01Exact/')):
                        os.mkdir(files[index].replace(Replace_pattern,'01Exact/'))
                    df = pd.DataFrame({'FNRATE':fnrate_array})
                    df.to_csv(files[index].replace(Replace_pattern,'01Exact/'+type_name+'.csv'))
                else:
                    if not os.path.exists(files[index].replace(Replace_pattern,'02Normal/')):
                        os.mkdir(files[index].replace(Replace_pattern,'02Normal/'))
                    df = pd.DataFrame({'FNRATE':fnrate_array})
                    df.to_csv(files[index].replace(Replace_pattern,'02Normal/'+type_name+'.csv'))
                    
            print('=========================================='+str(count)+'=====================================\n')
```

```python
tr_cnt = 0
normal_cnt = 0
exact_cnt =0

for i in range(0,len(Excel_files_list)):
    for j in range(0,len(Excel_files_list[i])):
        path = Excel_files_list[i][j]
        type_name = path.split('/')[-1]
        Replace_pattern = path.split('/')[-2]+'/'+path.split('/')[-1]
        if 'exact' in path:
            exact_cnt+=1
            if not os.path.exists(path.replace(Replace_pattern,'01Exact/')):
                os.mkdir(path.replace(Replace_pattern,'01Exact/'))
            os.system('cat '+path+' > '+path.replace(Replace_pattern,'01Exact/')+type_name)
        elif 'True' in path:
            tr_cnt+=1
            if not os.path.exists(path.replace(Replace_pattern,'03True/')):
                os.mkdir(path.replace(Replace_pattern,'03True/'))
            os.system('cat '+path+' > '+path.replace(Replace_pattern,'03True/')+type_name)
        else:
            normal_cnt+=1
            if not os.path.exists(path.replace(Replace_pattern,'02Normal/')):
                os.mkdir(path.replace(Replace_pattern,'02Normal/'))
            os.system('cat '+path+' > '+path.replace(Replace_pattern,'02Normal/')+type_name)
            
print(tr_cnt,normal_cnt,exact_cnt)
```
